```python
import pandas as pd
import numpy as np
from ctgan import CTGAN
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from scipy.stats import ks_2samp, chi2_contingency
import time
```

```python
df = pd.read_csv("com_salary.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
orig_df = df
```

```python
# Define preprocessing for numerical columns: impute missing values then scale
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Define preprocessing for categorical columns: impute missing values then encode
cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('encoder', OneHotEncoder(handle_unknown='ignore'))
])

# Combine both pipelines into a single column transformer
preprocessor = ColumnTransformer([
    ('num', num_pipeline, orig_df.select_dtypes(include=['int64', 'float64']).columns),
    ('cat', cat_pipeline, orig_df.select_dtypes(include=['object', 'category']).columns)
])
```

```python
# Apply the preprocessing steps to the original dataset
prepared_data = preprocessor.fit_transform(orig_df)

# Convert the prepared data back to a DataFrame for CTGAN
prepared_df = pd.DataFrame(prepared_data, columns=[*orig_df.select_dtypes(include=['int64', 'float64']).columns.tolist(), *preprocessor.named_transformers_['cat']['encoder'].get_feature_names_out(orig_df.select_dtypes(include=['object', 'category']).columns).tolist()])
```
